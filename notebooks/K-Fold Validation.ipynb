```python
import pandas as pd

from utils.preprocess import encode, load_and_evaluate_model

from sklearn.preprocessing import Normalizer
from sklearn.model_selection import train_test_split

import warnings
warnings.simplefilter('ignore')
```

```python
datapath = '../data/'

df = pd.read_csv(datapath+'kddcup.data.csv')
df
```

```python
X = df.drop(columns=['Y'])
Y = df['Y']

scaler = Normalizer().fit(X)
X = scaler.transform(X)
```

```python
df = pd.DataFrame(X, columns=df.drop(columns=['Y']).columns)
df['Y'] = Y
df
```

```python
df_80,fold_1=train_test_split(df,test_size=0.2,stratify=df['Y'],random_state=42)
```

```python
df_80['Y'].value_counts()/len(df_80)
```

```python
fold_1['Y'].value_counts()/len(fold_1)
```

```python
df_40,df_40_b=train_test_split(df_80,test_size=0.5,stratify=df_80['Y'],random_state=42)
fold_2,fold_3=train_test_split(df_40,test_size=0.5,stratify=df_40['Y'],random_state=42)
fold_4,fold_5=train_test_split(df_40_b,test_size=0.5,stratify=df_40_b['Y'],random_state=42)
```

```python
folds=[fold_1,fold_2,fold_3,fold_4,fold_5]
```

```python
for fold in folds:
    X, y = fold.drop(columns=['Y']), fold['Y']
    load_and_evaluate_model('tanh_dnn.hdf5', X, y)
```

```python
for fold in folds:
    X, y = fold.drop(columns=['Y']), fold['Y']
    load_and_evaluate_model('tanh_cnn.hdf5', X, y)
```
